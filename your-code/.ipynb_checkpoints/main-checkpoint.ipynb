## Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

1. Import libraries.

1. Define the initial API endpoint URL.

1. Make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.

1. Find out the total page count in the 1st page data.

1. Use a FOR loop to make requests for the additional pages from 2 to page count. Append the data of each additional page to the flatterned data object.

1. Print and review the data you obtained.

```python
# your code here
import requests
import pandas
response = requests.get(url)
results = response.json()

def all_chars():
    list_chars = pandas.DataFrame()
    for i in range(2): 
        url = 'https://rickandmortyapi.com/api/character/?page='
        if i == 0:
            continue  # Evitando errores
        elif i >= 1:
            new_url = str(url)+str(i)  # crea el url añadiendo el número de página
            new_request = requests.get(new_url)  # hace el request
            new_result = new_request.json()  # devuelve el dato solicitado en un JSON
            data = pandas.DataFrame(new_result['results']) # Creas el Dataframe
            list_chars = list_chars.append(data)  # Se van uniendo los dataframes
    #list_chars.to_csv('C:\Users\fpedrero\PycharmProjects\lab-parsing-api\your-code\chars.csv') # GIT_IGNORE
    return list_chars

#print(all_chars())
```

```python
df_RickMorty_chars = all_chars()
#df_RickMorty_chars
df_RickMorty_chars = df_RickMorty_chars.drop(columns=['type','created','url','image'])
df_RickMorty_chars
```
